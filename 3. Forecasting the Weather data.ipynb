#### Forecasting Through SQL Query

```python

%sql
WITH cte AS (
    SELECT *
    FROM (
        SELECT *,
        ROW_NUMBER() OVER (PARTITION BY id ORDER BY date DESC) AS row_num
        FROM forcasted_data
    ) t
    WHERE row_num <= 5
)
SELECT
    id,
    city_name,
    lon,
    lat,
    MAX(date),
    DATEADD(hour,1,max(time)),
    ROUND(AVG(temp), 2),
    temp_min,
    temp_max,
    pressure,
    humidity,
    dt,
    timezone,
    visibility,
    wind_deg,
    wind_speed,
    wind_gust,
    clouds_all,
    'True' AS is_forecasted_data
FROM cte
WHERE time = (SELECT MAX(time) FROM cte)
GROUP BY
    id,
    city_name,
    lon,
    lat,
    temp_min,
    temp_max,
    pressure,
    humidity,
    dt,
    timezone,
    visibility,
    wind_deg,
    wind_speed,
    wind_gust,
    clouds_all
```

####Forecasting the last  5 hours Average data through DataFrame

```python
def forecast_temp_data(df):
    from pyspark.sql import functions as F
    from pyspark.sql.window import Window
   
    
    # Create a Window specification for the row_number function
    window_spec = Window.partitionBy('id').orderBy(F.desc('date'))
    
    # Create the cte DataFrame with row_number
    cte = df.withColumn('row_num', F.row_number().over(window_spec)).where('row_num <= 5')
    
    # Perform the aggregation and select statements
    result = cte.groupBy('id') \
        .agg(
            F.first('city_name').alias('city_name'),
            F.first('lon').alias('lon'),
            F.first('lat').alias('lat'),
            F.max('date').alias('date'),
            F.expr("DATEADD(hour, 1, max(time))").alias('time'),
            F.round(F.avg('temp').cast('double'), 2).alias('temp'),
            F.first('temp_min').alias('temp_min'),
            F.first('temp_max').alias('temp_max'),
            F.first('pressure').alias('pressure'),
            F.first('humidity').alias('humidity'),
            F.first('dt').alias('dt'),
            F.first('timezone').alias('timezone'),
            F.first('visibility').alias('visibility'),
            F.first('wind_deg').alias('wind_deg'),
            F.first('wind_speed').alias('wind_speed'),
            F.first('wind_gust').alias('wind_gust'),
            F.first('clouds_all').alias('clouds_all'),
            F.lit('True').alias('is_forecasted_data')
        )
    
    return result.orderBy(F.col('date').desc())



```

```python
#Call function
result_df = forecast_temp_data(df)
result_df.display()
```

####Adding forecasted data into the Source Df

```python
final_df = df.union(result_df)

# Display the combined DataFrame
final_df=final_df.orderBy(col('time').desc())
final_df.display()
final_df.write.format("csv").option("header", "true").mode("overwrite").save("dbfs:/FileStore/shared_uploads/ujjwal.baral3980@kistcollege.edu.np/Weather_data__1_.csv")

```
