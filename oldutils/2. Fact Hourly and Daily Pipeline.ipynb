####Join Tables and select Column

```
from pyspark.sql.functions import hour,col
#date_df is in Dim Date & Dim Time table notebook


Join_df=df.join(date_df,date_df['fullDate']==df['date'],'inner') 
Join_df = Join_df.join(time_df, hour(Join_df['date']) == time_df['timeID'], 'inner')



# Select multiple columns
selected_columns = ["dateID","timeID","city_id", "dt","temp","temp_min","temp_max","visibility","pressure","humidity","wind_deg","wind_gust","created_on"]
fact_df = Join_df.select(*selected_columns)


fact_df.orderBy(col("created_on").desc()).display()
fact_df.write.format('delta') \
    .mode('overwrite') \
    .option("mergeSchema", "true") \
    .saveAsTable('silver_processed_fact_df')
```

####Show latest Hourly Data to the fact Table and Update in log table

```
@keep_log
def hourly_fact():
    from datetime import datetime
    df = fact_df.dropDuplicates(['city_id','timeID']).orderBy(col("created_on").desc())
    start = datetime.fromtimestamp(df.selectExpr("min(dt)").first()[0])
    end = datetime.fromtimestamp(df.selectExpr("max(dt)").first()[0])
    return df,start,end

```

```
hourly_df= hourly_fact('Hourly fact Load','Fact_hourly_table')
hourly_df.display()
```

####Show Daily Table and Update in log Table

```
@keep_log
def daily_data(fact_df):
    from pyspark.sql.window import Window
    from pyspark.sql.functions import avg
    from datetime import datetime

    window_spec = Window.partitionBy(fact_df.dateID)
    avg_weather_data = fact_df.withColumn("avg_temperature", avg(fact_df.temp).over(window_spec)) \
                                .withColumn("avg_temp_min", avg(fact_df.temp_min).over(window_spec)) \
                                .withColumn("avg_temp_max", avg(fact_df.temp_max).over(window_spec)) \
                                .withColumn("avg_pressure", avg(fact_df.pressure).over(window_spec)) \
                                .withColumn("avg_visibility", avg(fact_df.visibility).over(window_spec)) \
                                .withColumn("avg_humidity", avg(fact_df.humidity).over(window_spec)) \
                                .withColumn("avg_wind_deg", avg(fact_df.wind_deg).over(window_spec)) \
                                .withColumn("avg_wind_gust", avg(fact_df.wind_gust).over(window_spec))
    start = datetime.fromtimestamp(fact_df.selectExpr("min(dt)").first()[0])
    end = datetime.fromtimestamp(fact_df.selectExpr("max(dt)").first()[0])
    return avg_weather_data, start, end


```

```
df = daily_data('Daily load data', 'daily_fact', fact_df)
df.select("load_run_id", "created_by", "avg_temperature", "avg_temp_min", "avg_temp_max", 
          "avg_pressure", "avg_visibility", "avg_humidity", "avg_wind_deg", "avg_wind_gust", 
          "dateID", "timeID", "city_id").display()
```
